```python
import pandas as pd

df = pd.read_csv('/home/3a266a5e-f8a6-4c1f-ab39-f0a241d712b8/marketing campaign /marketing_campaign.csv', sep=';')
df.head()
```

```python
# How many rows and columns do we have?
print(df.shape)

# What are all the column names?
print(df.columns.tolist())

# Basic stats on the data
df.describe()
```

```python
# Check missing values
print(df.isnull().sum())
```

```python
# Check income distribution
import matplotlib.pyplot as plt

df['Income'].hist(bins=50)
plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Count')
plt.show()
```

```python
# Remove missing income values
df = df.dropna(subset=['Income'])

# Remove income outlier (anyone making over $200,000 is likely bad data)
df = df[df['Income'] < 200000]

# Confirm how many rows we have now
print(f'Clean dataset has {df.shape[0]} customers')
```

```python
# Create customer age column from Year_Birth
import datetime
current_year = datetime.datetime.now().year
df['Age'] = current_year - df['Year_Birth']

# Create total spending column
df['TotalSpend'] = (df['MntWines'] + df['MntFruits'] + 
                    df['MntMeatProducts'] + df['MntFishProducts'] + 
                    df['MntSweetProducts'] + df['MntGoldProds'])

print(df[['Age', 'TotalSpend']].describe())
```

```python
import matplotlib.pyplot as plt

spend_by_education = df.groupby('Education')['TotalSpend'].mean().sort_values(ascending=False)

spend_by_education.plot(kind='bar', color='steelblue', figsize=(10,5))
plt.title('Average Total Spend by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Average Spend ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
spend_by_marital = df.groupby('Marital_Status')['TotalSpend'].mean().sort_values(ascending=False)

spend_by_marital.plot(kind='bar', color='coral', figsize=(10,5))
plt.title('Average Total Spend by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Average Spend ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.scatter(df['Income'], df['TotalSpend'], alpha=0.3, color='green')
plt.title('Income vs Total Spend')
plt.xlabel('Income ($)')
plt.ylabel('Total Spend ($)')
plt.show()
```

```python
correlation = df['Income'].corr(df['TotalSpend'])
print(f'Correlation between Income and Total Spend: {correlation:.2f}')
```

```python
# Which campaign had the best response rate?
campaigns = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 
             'AcceptedCmp4', 'AcceptedCmp5']

response_rates = df[campaigns].mean() * 100

response_rates.plot(kind='bar', color='purple', figsize=(10,5))
plt.title('Response Rate by Campaign (%)')
plt.xlabel('Campaign')
plt.ylabel('Response Rate (%)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

print(response_rates)
```

```python
print("=" * 50)
print("MARKETING CAMPAIGN ANALYSIS - KEY FINDINGS")
print("=" * 50)
print(f"Total Customers Analyzed: {df.shape[0]}")
print(f"Income vs Spend Correlation: {correlation:.2f}")
print(f"\nBest Performing Campaign: {response_rates.idxmax()}")
print(f"Best Campaign Response Rate: {response_rates.max():.1f}%")
print(f"\nTop Spending Education Level: {spend_by_education.idxmax()}")
print(f"Top Spending Marital Status: {spend_by_marital.idxmax()}")
```
